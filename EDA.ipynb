```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("./vehicles_us.csv")

df.head()
df.info()
df.isna().sum()
df.describe(include="all")
```

**Clean just enough for plotting**

```python
# Basic cleaning for charts
df_eda = df.copy()

# Ensure numeric
df_eda["price"] = pd.to_numeric(df_eda["price"], errors="coerce")
df_eda["odometer"] = pd.to_numeric(df_eda["odometer"], errors="coerce")
df_eda["model_year"] = pd.to_numeric(df_eda["model_year"], errors="coerce")

# Drop clearly bad rows for plotting
df_eda = df_eda.dropna(subset=["price"])
df_eda = df_eda[df_eda["price"] > 0]
```

**Plot 1 histogram**

```python
fig1 = px.histogram(df_eda, x="price", nbins=60, title="Price distribution (price > 0)")
fig1.show()
```

**Plot 2 histogram**

```python
fig2 = px.histogram(df_eda.dropna(subset=["odometer"]), x="odometer", nbins=60, title="Odometer distribution")
fig2.show()
```

**Scatter plot**

```python
df_scatter = df_eda.dropna(subset=["odometer"])
fig3 = px.scatter(df_scatter, x="odometer", y="price", title="Price vs Odometer")
fig3.show()
```

**Another scatter**

```python
df_scatter2 = df_eda.dropna(subset=["model_year"])
fig4 = px.scatter(df_scatter2, x="model_year", y="price", color="condition", title="Price vs Model Year by Condition")
fig4.show()
```
